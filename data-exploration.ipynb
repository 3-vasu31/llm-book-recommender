```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ax=plt.axes()
sns.heatmap(books.isna().transpose(),cbar=False,ax=ax)

plt.xlabel("Columns")
plt.ylabel("Missing values")

plt.show()
```

```python
import numpy as np
books["missing_description"]=np.where(books["description"].isna(),1,0)# put 1 if missing else 0
books["age_of_book"]=2024-books["published_year"]
```

```python
columns_of_interest=["num_pages","age_of_book","missing_description","average_rating"]


#use "pearson" when linear relationship, use "spearman" when the relationship is monotonic but not linear
correlation_matrix=books[columns_of_interest].corr(method="spearman")
sns.set_theme(style="white")
plt.figure(figsize=(10,10))
heatmap=sns.heatmap(correlation_matrix,annot=True,fmt=".2f",cmap="coolwarm"
                    ,cbar_kws={"label":"Spearman correlation"})

heatmap.set_title("Correlation heatmap")
plt.show()
```

```python
# we want to delete all those rows which have missing description, num_pages, average_rating, or published_year
books[(books["description"].isna())|
      (books["num_pages"].isna())|
      (books["average_rating"].isna())|
      (books["published_year"].isna())] 
```

```python
# negating the above condition to get the rows which have all the values
# storing in new dataframe
book_missing=books[~(books["description"].isna())&
      ~(books["num_pages"].isna())&
      ~(books["average_rating"].isna())&
      ~(books["published_year"].isna())] 
```

```python
book_missing["words_in_description"]=book_missing["description"].str.split().str.len()
```

```python
# removing the rows with description length is less than 25
book_missing_25_words=book_missing[book_missing["words_in_description"]>=25]
```

```python
book_missing_25_words
```

## Looking at the 'subtitle' field
subtitle field is short that it can't give description but we can add description and subtitle fields with ";".

```python
# when subtilte is missing, use title as title and when it is not missing, aggrigate title and subtitle 
book_missing_25_words["title_and_subtitle"]=(
    np.where(book_missing_25_words["subtitle"].isna(),book_missing_25_words["title"],
            book_missing_25_words[["title","subtitle"]].astype(str).agg(": ".join,axis=1) )
)
```

```python
book_missing_25_words
```

```python
# we now want a unique identifier with each description
book_missing_25_words["tagged_description"]=book_missing_25_words[["isbn13","description"]].astype(str).agg(" ".join,axis=1)
```

```python
book_missing_25_words["tagged_description"]
```

```python
# deleting the columns which are not needed and saving the file

{
    book_missing_25_words.drop(
        ["subtitle",
         "missing_description",
         "age_of_book",
         "words_in_description"], axis=1)
         .to_csv("books_cleaned.csv",index=False)
}
```
